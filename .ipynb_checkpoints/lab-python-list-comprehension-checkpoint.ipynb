```python
 products = ["t-shirt", "mug", "hat", "book", "keychain"]
customer_orders = set()

## 1
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    print(f"Current inventory: {inventory}")
    return inventory
## 2
def get_customer_orders():
    inventory = initialize_inventory(products)
    print(" ")
    num_orders = int(input("Enter the number of customer orders: "))
    valid_inputs = []
    while len(valid_inputs)<num_orders:
        order = input("Enter the name of a product a customer wants to order: ").strip()
        if order in inventory.keys() :
            customer_orders.add(order)
            valid_inputs.append(order)
            inventory[order]=inventory[order]-1
            inventory = update_inventory(inventory)
        else:
            print("""Please make sure you entered your order correctly!
                     Product might be out of stock.""" )
    return customer_orders, inventory
## 4
def update_inventory(inventory):
    global updated_inventory
    updated_inventory={product : value for product, value in inventory.items() if value > 0}
    return updated_inventory
    
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage = len(customer_orders)/len(products)*100
    return total_products_ordered, percentage

def print_order_statistics(calculate_order_statistics):
    total_products_ordered, percentage = calculate_order_statistics(customer_orders, products)
    print(" ")
    print(f"You ordered {customer_orders}")
    print(" ")
    print("Order statistics:")
    print(f"Total products ordered: {total_products_ordered}")
    print(f"Percentage of unique products ordered: {percentage}% ")

def print_updated_inventory(inventory):
    print(" ")
    print("Updated inventory: ")
    print(" ")
    for key, value in inventory.items():
        print(f"{key}: {value}")
## 3
def total_cost(customer_orders):
    print("")
    total_cost = sum(float(input(f"Enter the price of {product}:€ "))for product in customer_orders)
    print(f"Total Price:€ {total_cost}") ## 5
    return total_cost      
        
        
def start_shop():
    get_customer_orders()
    print_order_statistics(calculate_order_statistics)
    print_updated_inventory(updated_inventory)
    total_cost(customer_orders)
    
start_shop()
```
